# Phase 1: Liver Disease Dataset Preprocessing
This notebook loads the raw ILPD dataset, assigns headers, handles missing values, and encodes categorical variables.

```python
import pandas as pd
import numpy as np
df = pd.read_csv('/Users/ishaanjain/DL PROJECT/data_raw/Indian Liver Patient Dataset (ILPD).csv', header=None)
df
```

```python
# Assigning descriptive column names
df.columns = [
    "age", "gender", "total_bilirubin", "direct_bilirubin",
    "alkaline_phosphotase", "alt", "ast", "total_proteins",
    "albumin", "albumin_globulin_ratio", "liver_disease"
]

df
```

```python
# Encoding gender: Male to 1, Female to 0
df["gender"] = df["gender"].map({"Male": 1, "Female": 0})

# Converting target label: Originally-> 1 (disease) and 2 (no disease) 
# Mapping to 1 (disease) and 0 (no disease) for standard ML use
df["liver_disease"] = df["liver_disease"].map({1: 1, 2: 0})

print("Encoding complete.")
```

```python
df
```

```python
df.isnull().sum() #before
```

```python
# Albumin_globulin_ratio has missing entries; we fill them with the median
df.fillna(df["albumin_globulin_ratio"].median(), inplace=True)

print(f"Missing values remaining: {df.isnull().sum().sum()}")
```

```python
df.isnull().sum() #after
```

```python
# Save the cleaned dataset for Phase 2
df.to_csv("ilpd_cleaned.csv", index=False)
print("Cleaned dataset saved as 'ilpd_cleaned.csv'")
df.head()
```
